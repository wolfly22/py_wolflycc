```python
import psutil
```

```python
nets = psutil.net_connections()
for i in nets:
    ip = "IP4" if i.family==2 else "IP6"
    ty = "TCP" if i.type==1 else "UDP"
    prog = psutil.Process(pid=i.pid)
    print(f"{ip}:{i.family}    {ty}  {i.laddr.ip:>25s}:{i.laddr.port}      {prog.name()}")
```

for n in nets:
    ip = "IP4" if n.family==2 else "IP6"
    ty = "TCP" if n.type==1 else "UDP"
    print(f"{ip}    {ty}     {n.laddr.ip:15s}:{n.laddr.port}   pid:{n.pid}")

```python
wins = psutil.win_service_iter()
for i in wins:
    print(f"{i.name():25s}{i.display_name():30s}{i.binpath()}   {i.pid()}")
```

```python
pids = psutil.pids()
```

```python
for pid in pids:
    if pid ==0:
        continue
    p = psutil.Process(pid)
    print(f"{p.name():100s}{p.exe()}")
    conn = p.connections()
    print(f"   connections_num:{len(conn)}")
    for c in conn:
        #print(c)
        ty == "TCP" if c.type==1 else "UDP"
        ra = c.raddr
        if ra:
            print(f"   {ty}  {c.laddr.ip}:{c.laddr.port} {c.raddr}:{c.raddr}")
        else:
            print(f"   {ty}  {c.laddr.ip}:{c.laddr.port} ")
```
