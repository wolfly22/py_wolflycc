```python
import tushare as tu
import pandas as pd
import os
def buy2sale(code_list):
    for code_i in code_list:
        pds=pd.DataFrame()
        try:
            pds=tu.get_today_ticks(code_i)
            str_file_name= code_i + '.xlsx'
            file_name=os.path.join('E:\\gu\\202109\\',str_file_name)
            print(file_name)
            pds.to_excel(file_name)
            print('Done!!!')
        except OSError:
            print('%s 数据无法下载！'  %(code_i))
            continue
```
